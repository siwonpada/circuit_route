```python
from qiskit.dagcircuit import DAGCircuit
import qiskit.qasm2
from qiskit.converters import circuit_to_dag
```

```python
circuit = qiskit.qasm2.load('../data/3_17_13.qasm')

dag = circuit_to_dag(circuit)
dag.draw()
```

```python
print(dag.two_qubit_ops())
```

```python
nodes = dag.op_nodes()

print(nodes[1].name, nodes[1].qargs, nodes[1].cargs)
for i in dag.op_predecessors(nodes[1]):
    print(i.name, i.qargs, i.cargs)
```

operation을 추가하는 방법은 단 두가지 있다. 앞에 하나를 추가하거나 뒤에 하나를 추가하거나 우리는 뒤에 하나를 추가하는 방법을 사용해야한다. [DAGCircuit.apply_operation_back](https://docs.quantum.ibm.com/api/qiskit/qiskit.dagcircuit.DAGCircuit#apply_operation_back)

```python
from qiskit.circuit.library.standard_gates import SwapGate
qubits = dag.qubits
out =dag.output_map[qubits[0]]

dag.apply_operation_back(SwapGate(), [qubits[0], qubits[1]])


print(dag.output_map)
```
